```python
#set up the libraries
# we use apriori for SUPPORT and association_rules for CONFIDENCE, LIFT and CONVICTION
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#Loading the dataset
df=pd.read_csv('dataset1.csv')
df.head()
```

```python
df.isnull().sum()
df.loc[(df!=0).any(axis=1)]
```

```python
#Since it's missing a column header, so we create new column header name
df=pd.read_csv('dataset1.csv', names=['Groceries'], sep=',')
df.head(10)
```

```python
#data size
df.shape
```

```python
#we are turning the pandas dataframe into a list of lists
data=list(df["Groceries"].apply(lambda x:x.split(",")))
data
```

```python
#Let's transform the list with one-hot encoding
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)
df=df.replace(True,1)
df=df.replace(False,0)
df
```

```python
#Set a threshold value for the support value and calculate the support value
df=apriori(df,min_support=0.1, use_colnames=True)
df
```

Any item or combinations less than 1.0 are not shown.

```python
#Here is the CONFIDENCE (Try different min_threshold value to explore this more)
df_ar=association_rules(df,metric="confidence", min_threshold=0.3)
df_ar
```

The first two rows (Apple Juice -> Baby Food) shows a confidence of 1.0 which means the two items are bought together 100%. The third row (Eggs -> Kidney beans) shows a confidence of .57 which means these two items are bought together only 57% of the time.

```python
#Here is LIFT (try different min_threshold value to explore more) 
df_ar=association_rules(df,metric="lift", min_threshold=0.8)
df_ar
```


The first row (Apple Juice & Baby food) show a LIFT of 3.67 which is a positive correlation between the two items. 

```python
#Let's view our interpretation values using the Association rule function
df_ar=association_rules(df,metric="conviction", min_threshold=0.8)
df_ar
```

Conviction is similar to LIFT however Conviction will give you a directed measure meaning that on row 3 (Kidney Beans-> Eggs) has a higher conviction than (Eggs -> Kidney Beans) thus it is more likely that the first instance (Kidney Beans -> Eggs) will lead to a purchase.
